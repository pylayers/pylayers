Create the simulation object with `WHERE1.ini` layout. This layout corresponds to the office building where the first WHERE1 UWB measurement campaign has been conducted.  

The layout .ini file do not contain the whole necessary information required for further calculation. In particular, the different graph are not built. 
It is necessary to build those graph before gohing ahead. This calculation takes a while thus the post-processed graph information is stored in various 
gpickle file. If those files have been already calculated the dumpr method read those data, otherwise the build method is invoqued and the graphs are saved. 

```python
S = Simul()
# loading a layout 
filestr = 'WHERE1_clean'
S.layout(filestr+'.ini','matDB.ini','slabDB.ini')
try:
    S.L.dumpr()
except:
    S.L.build()
    S.L.dumpw()
```

# How to use the GrRayTud object? 

## Gt is an object from class GrRayTud (Group of rays in TUD format) 

### Gt.eval
Thhis method evaluates the field along the rays. 

```python
Gt.eval()
```

### Gt.info(r)

returns the information associated to a given ray number r

```python
Gt.info(7)
```

### Gt.ray(r)

return the index of the interactions of ray r into the Gt.I.I matrix 

```python
Gt.ray(0)
```

### Gt.typ(r)

return the type of the interactions of ray r 

```python
Gt.typ(0)
```

## Gt Attributes

### Gt.I

Gt has an attribute I, which is a class Interactions.

This class Interactions gather all basis interactions : B / T / R / L 

```python
print Gt.I.B
print Gt.I.R
print Gt.I.T
print Gt.I.L
```

All basis interactions B / T / R / L have the same attributes:

* idx : the index of the interaction 
* data : which is a np.shape(self.data) = len(self.idx) , 3 
* data[:,0] = theta
* data[:,1] = si
* data[:,2] = sr or st ( named sout in the following)

```python
print Gt.I.R.idx,'\n'
print Gt.I.R.data
```

T and R basis interractions have also an extra attribute:

* dusl : a dictonnary of used slab ( the number is the **position** into the self.idx, not direcly the index)


```python
print Gt.I.R.dusl
```
