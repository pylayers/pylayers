# Initialization

```python
from pylayers.antprop.coverage import *
import time
```

Instantiate a coverage object.
By defaut, the TA-Office.str layout strucure is loaded.

This information can be modified into the **coverage.ini** file into the project directory.

```python
C=Coverage()
C.L.filename
```

# Coverage

Coverage required Layout graphs.

- If the graphs for the given layout have already been built, they are loaded from the project directory. 
- Otherwise, they are being built  **(it may take few minutes )** and stored into the project directory for the next execution.


```python
try :
    C.L.dumpr()
except:
    C.L.build()
```

Now, you can start the coverage operation :

```python
C.cover()
```

You can display the coverage result with 

```python
C.showPr()
```

By default, The transmitter is set at 

```python
C.tx
```

This can be modified, and the coverage can be updated

```python
C.tx = np.array((30,12))
C.cover()
C.showPr()
```

Quite fast, for any Tx position

```python
C.tx = np.array((39,1))
start = time.time()
C.cover()
finish = time.time()
C.showPr()
print 'in %1.2f seconds' % (finish-start)
```
