Reading and saving in vsh3 format 

```python
for name in serie1:
    filein = name+'.mat'
    A = Antenna(filein,'ant/UWBAN/Matfile')
    A = vsh(A)
    A.C.s1tos2(35)
    A.C.s2tos3(1e-5)
    eth,eph,err = A.errel(20,dsf=1,typ='s3')
    print name,err,len(A.C.Br.ind3)
    A.savevsh3()
```
