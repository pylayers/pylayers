The cell below seeks for all the normals of the cycle using the Lr2n (LinearRing to Normal) function and determine those segments which are on the front line. 

If $\hat{n_j}.\vec{v}<0$ then segment $j$ is on the front line



```python
def frontline(L,nc,v):
    """ determine cycle frontline
    
    Parameters
    ----------
    L : Layout 
    nc : cycel number
    v : direction vector 
    
    """
    npt = filter(lambda x: x<0, L.Gt.node[nc]['cycle'].cycle)
    nseg = filter(lambda x: x>0, L.Gt.node[nc]['cycle'].cycle)
    pt  = map(lambda npt : [L.Gs.pos[npt][0],L.Gs.pos[npt][1]],npt)
    pt1 = np.array(pt)
    n1 = Lr2n(pt1.T)
    ps = sum(n1*v,axis=0)
    u = where(ps<0)[0]
    nsegf = map(lambda n: nseg[n],u)
    return nsegf


```

```python
fig=gcf()
ax=fig.gca()
nc = 5

v = np.array([2,2]).reshape(2,1)
pol  = L.Gt.node[nc]['polyg']
fig,ax = pol.plot(fig=fig,ax=ax)
nsegf = frontline(L,nc,v)
fig,ax = L.plot_segments(nsegf,fig=fig,ax=ax,linewidth=3,color='red')
```
